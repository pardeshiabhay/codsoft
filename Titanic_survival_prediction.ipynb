```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('./downloads/archive/tested.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df
```

```python
new_df = df.drop(['PassengerId','Name','Ticket'],axis =1)
new_df
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
new_df['Sex'].unique()
```

```python
new_df['Sex'] = new_df['Sex'].map({'male':0,'female':1})
new_df
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le = LabelEncoder()
new_df['Cabin'] = le.fit_transform(new_df['Cabin'])
new_df.info()
```

```python
new_df
```

```python
emb_cat = new_df['Embarked'].value_counts()
emb_cat.index
```

```python
most_occuring_embarked = emb_cat.index[0]
new_df['Embarked'] = new_df['Embarked'].fillna(most_occuring_embarked)

emb = pd.get_dummies(new_df['Embarked'],drop_first = False, prefix = 'Embarked').astype("bool")
emb
```

```python
new_df = pd.concat([new_df,emb],axis = 1 )


new_df.drop('Embarked',axis = 1, inplace = True)
new_df
```

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
new_df['Fare'] = new_df['Fare'].fillna(new_df['Fare'].mean())
new_df.isnull().sum()
```

```python
new_df.info()
```

```python
y = new_df['Survived']
X = new_df.drop('Survived',axis =1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()


logreg.fit(X_train,y_train)
```

```python
pred = logreg.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred) * 100
```

```python
import pickle

pickle_out = open('model.pkl', 'wb')
pickle.dump(logreg, pickle_out)
pickle_out.close()
```
